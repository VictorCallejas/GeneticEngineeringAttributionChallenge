```python
# this determines what results are saved. This cutoff is generous
cutoff = 10
result_name = "test_set_blast_output2"
!blastn -db train.fasta -query query.fasta -out {result_name}.csv -outfmt 6 -max_target_seqs 9999999 -evalue {cutoff} -max_hsps=1 -num_threads 8
```

```python
# now we will get the training set labels to match sequences hits
# back to labs
# you'll need to download this yourself
train_y = pd.read_csv('train_labels.csv')
```

```python
# confirm that ids match
(train.sequence_id == train_y.sequence_id).all()
```

```python
# looks good, so we can just rely on the row index
train_y = train_y.drop('sequence_id', axis=1)
```

```python
test_len = len(test)
```

```python
!head {result_name + '.csv'}
```

```python
predictions = []
k = 10
headers = ['query_id', 'subject_id', 'identity', 'alignment length', 'mismatches', 'gap opens', 'q. start', 'q. end', 's. start', 's. end', 'evalue', 'bit score']
result = vpd.from_csv(result_name, sep='\t', names=headers, index_col=False, copy_index=True, convert='True')
```

```python
result.head()
```

# This is where most of the algorithm lives.

For each element in the test set, parse all the "hits" by query_id, then get the top k subject_ids and look up the corresponding training-set labels. Avg these into the vector format.

```python

# for every element in query
for i in range(test_len):
    if i % 100 == 0:
        print(i)
    queryi_result = result[result["query_id"] == i].to_pandas_df()
    k_best_idxs = np.asarray(queryi_result.sort_values("evalue").loc[:,"subject_id"].values[:k], dtype=int)
    # get the top k match's lab vectors, and take the mean along the "hits" axis to get a vector 
    # which is 0 everywhere except 1/k in the topk lab matches from training set
    predictions.append(np.mean(train_y.iloc[k_best_idxs, :].values, axis=0))
```

```python
predict_df = pd.DataFrame(predictions)
```

```python
# The na's are cases where the query had no matches in the training set.
predict_df.shape[0] - predict_df.dropna().shape[0]
```

```python
sub_format = pd.read_csv("submission_format.csv")
sub_format
```

```python
# replace nans with the uniform probability vectors.
for i, row in predict_df.iterrows():
    if pd.isnull(row).any():
        predict_df.iloc[i,:] = np.ones((1314)) / 1314
```

```python
predict_df.shape[0] - predict_df.dropna().shape[0]
```

```python
# removed the nans, good. Now put in the submission format.

sub_format.iloc[:, 1:] = predict_df.values
```
